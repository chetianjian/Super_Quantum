```python
def remove_diag(arr: np.ndarray):
    return np.array(list(map(lambda i: np.delete(arr[i], i), range(len(arr)))))



def perfect_beta(df1, df2, bias=False):
    """
    :param df1: pd.DataFrame with shape n * d
    :param df2: pd.DataFrame with shape n * 1
    :param bias: bool with default by False, denote whether the estimate is biased (True) or unbiased (False)
    :return: pd.DataFrame with shape n * n
    """

    return remove_diag(np.cov(df1.T, df2.T, bias=bias)) / df2.var(ddof=bias)
```

```python
def ri_beta(ri, rm, bias=False):
    """
    :param ri: pd.DataFrame or np.ndarray with length n or shape n * 1
    :param rm: pd.DataFrame or np.ndarray with shape n * 1
    :param bias: bool with default by False, denote whether the estimate is biased (True) or unbiased (False)
    :return: pd.DataFrame with shape n * n
    """
    try:
        cov_mat = np.cov(ri, rm, bias=bias)
        return cov_mat[0, 1] / cov_mat[1, 1]
    except:
        return np.nan
```

```python
def rp_beta(rp):
    arr = rp.index
    return rp.apply(lambda col: ri_beta(col, arr), axis=0)
```
